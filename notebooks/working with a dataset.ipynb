# Working with a dataset

First import `ghiaseddin` and use the `%matplotlib` magic to show the images inline

```python
import sys
sys.path.append('../')

import ghiaseddin
%matplotlib inline
```

Now you can easity create a dataset object of the zappos dataset.

If you don't have the files for the dataset you can easily use the following to download the dataset

```python
# %run ../ghiaseddin/scripts/download-dataset-zappos.py
```

```python
dataset = ghiaseddin.Zappos50K1(root=ghiaseddin.settings.zappos_root, attribute_index=0, split_index=0)
```

Here we can use the dataset object to view pairs from training and test data.

```python
dataset.show_pair(0)
```

We can also create a generator to do throw the dataset in batches of arbitrary size

```python
for b in dataset.test_generator(batch_size=128):
    print 'number of items in batch:', len(b) # the length of the batch
    for item in b:
        images_path, target = item # the generator returns a tuple (images_path, target)
        img1_path, img2_path = images_path # the images_path is itself a tuple (img1_path, img2_path)
        
        # we can print 
        print 'absolute path of the first image:', img1_path
        print 'absolute path of the second image:', img2_path
        print 'the target value:', target
        break
    break
```
